Converting the labels into numeric form.

classification machine learning proble; Is predicting if a datapoint belongs to one class or another. predicting if a person is diabetic or non diabetic.

```python
#Importing the dependencies

import pandas as pd
from sklearn.preprocessing import LabelEncoder  #labelEncoder function
```

Label Encoding of Breast CAncer Dataset

```python
cancer = pd.read_csv('breast_cancer_data.csv')
```

```python
#first 5 rows
cancer.head()
```

```python
#finding the count of differrnt labels

cancer['diagnosis'].value_counts()
```

Converting the 'B' and 'M to corresponing lables with numerical values

```python
#loading the label encoder function
#It labels values between 0 and 1

label_encoder = LabelEncoder()
```

```python
#transforming values of diagnosis to 0 nad 1
labels = label_encoder.fit_transform(cancer.diagnosis)
```

```python
#Appending the labels to the data frame.
#It will create a new column call target for the label encoding

cancer['target'] = labels
```

```python
cancer.head()
```

0 = Benign
1 = Malig nant

```python
cancer['target'].value_counts()
```

Label Encoding of Iris Data

```python
iris = pd.read_csv('iris_data.csv')
```

```python
iris.head()
```

Transforming the labels to numerical value

```python
iris['Species'].value_counts()
```

Loading the Label Encoder

```python
#Loading the Label Encoder
label_encode = LabelEncoder()
```

Creating the lables and Storing it in the Variable iris_label

```python
#Creating the lables and Storing it in the Variable iris_label

iris_labels = label_encode.fit_transform(iris.Species)
```

```python
#Appending the labels(new column called target) to our dataset
iris['target'] = iris_labels
```
