```python
import matplotlib.pyplot as plt
```

```python
#import numpy to get some value
import numpy as np
```

```python
x = np.linspace(0,10,100)   #x=angle
#this function give data, from the script it will give evenly space 100 values between 1 and 10
y = np.sin(x)    #y = sin
z = np.cos(x)    #z = cos
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('Angle (radians)')
plt.title('cosine wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g*')
plt.plot(x, np.cos(x) ,'r')
plt.show()
```

BAR PLOT

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#0 and 0 rpresents the cordinates ie the origin
#1 and 1 reps the rectagle
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]

ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

PIE CHART
Is used to find the distribution of the data

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])

languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]

ax.pie(people, labels = languages, autopct = '%1.1f%%')
plt.show()
```

SCATTER PLOT

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color = 'g')
ax.scatter(x,z, color = 'r')
plt.show()
```

3D SCATTER PLOT

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c =z, cmap = 'Blues')
plt.show()
```
